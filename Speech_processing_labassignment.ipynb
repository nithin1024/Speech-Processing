## Importing Speech file


```python
audio_file = "/content/LJ001-0006.wav"
y,sr = librosa.load(audio_file)
```

# Basic Information of the file

```python
print("sampling rate : ", sr)
print("No of Samples : ", y.shape)
print("Audio Duration (seconds) : ", librosa.get_duration(y=y, sr=sr))
```

# Basic operations and plots

```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import soundfile as sf
import simpleaudio as sa

sliced_audio = y[:int(2 * sr)]
normalized_audio = librosa.util.normalize(y)

plt.figure(figsize=(12, 6))

plt.subplot(3, 1, 1)
plt.plot(np.arange(len(y)) / sr, y)
plt.title("Original Audio")
plt.xlabel("Time (seconds)")

plt.subplot(3, 1, 2)
plt.plot(np.arange(len(sliced_audio)) / sr, sliced_audio)
plt.title("Sliced Audio (First 2 seconds)")
plt.xlabel("Time (seconds)")


plt.subplot(3, 1, 3)
plt.plot(np.arange(len(normalized_audio)) / sr, normalized_audio)
plt.title("Normalized Audio")
plt.xlabel("Time (seconds)")

plt.tight_layout()
plt.show()

amplified_audio = y * 1.5
deamplified_audio = y * 0.5
up_sampled_audio = librosa.resample(y, orig_sr=sr, target_sr=sr * 2)
down_sampled_audio = librosa.resample(y, orig_sr=sr, target_sr=sr // 2)

plt.figure(figsize=(12, 8))

plt.subplot(5, 1, 1)
plt.plot(np.arange(len(amplified_audio)) / sr, amplified_audio)
plt.title("Amplified Audio")
plt.xlabel("Time (seconds)")

plt.subplot(5, 1, 2)
plt.plot(np.arange(len(deamplified_audio)) / sr, deamplified_audio)
plt.title("De-amplified Audio")
plt.xlabel("Time (seconds)")

plt.subplot(5, 1, 3)
plt.plot(np.arange(len(up_sampled_audio)) / (sr * 2), up_sampled_audio)
plt.title("Up-sampled Audio")
plt.xlabel("Time (seconds)")


plt.subplot(5, 1, 4)
plt.plot(np.arange(len(down_sampled_audio)) / (sr // 2), down_sampled_audio)
plt.title("Down-sampled Audio")
plt.xlabel("Time (seconds)")


plt.tight_layout()
plt.show()
```
